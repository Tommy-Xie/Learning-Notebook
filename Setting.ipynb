# jupyter 使用conda创建的虚拟环境

## 生成此环境对应的jupyter lab选项卡

1.先在要应用的环境中安装ipykernel

pip install ipykernel



2.安装jupyter menus

python -m ipykernel install --user --name ancillary --display-name "Python (ancillary)"



Jupyter 在内部使用 --namevalue。这些命令将覆盖任何具有相同名称的现有内核。--display-name 是您在笔记本菜单中看到的内容。

3.选择环境

此时打开jypyter lab就能看到刚刚创建的环境对应的选项卡了
