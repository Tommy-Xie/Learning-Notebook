# install pycocotools 

在python命令行下输入pip install pycocotools可以直接安装上pycocotools 2.0，但是这是假象.虽然可以导入pycocotools，但是导入pycocotools._mask时就会出错，提示不存在这个module，所以需要重新安装，但是不是用pip一次又一次的重装，而是按照网上的方法，去github上面下载源码：
https://github.com/philferriere/cocoapi

下载到本地


下载并安装vc++ 2015 build tools，然后修改系统默认编译环境，在系统环境变量中将箭头所指的Microsoft Visual Studio 12.0修改为Microsoft Visual Studio 14.0。

因为12.0对应的是vs2013，编译出来的不能用，因此要修改刚才安装的vc++ 2015对应的编译环境。修改完以后，然后在pythonAPI这个文件夹下打开命令行窗口，输入python setup.py build_ext --inplace 进行编译源码，

然后再执行：

all:
    # install pycocotools locally
    python setup.py build_ext --inplace
    rm -rf build

install:
    # install pycocotools to the Python site-packages
    python setup.py build_ext install
    rm -rf build

这样就会将编译好的pycocotools安装到python对应的sitepakages中：

验证：

import pycocotools
